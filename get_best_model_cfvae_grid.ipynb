```python
import os
import pandas as pd
import glob
import yaml
```

```python
data_path = 'data'
performance_path = os.path.join(data_path, 'performance', 'cfvae_grid')
```

```python
def path_split(x, num_iters = 1):
    for i in range(num_iters - 1):
        x = os.path.split(x)[0]
    return os.path.split(x)[1]
```

```python
files_eval = glob.glob(os.path.join(performance_path, '**', '*_eval.csv'), recursive=True)
df_eval = pd.concat({(file, path_split(file, 4), path_split(file, 3), path_split(file, 2)): 
                pd.read_csv(file) for file in files_eval})
df_eval = df_eval.rename_axis(index = ['file_path', 'outcome', 'sensitive_variable', 'experiment_id', 'index'])
```

```python
df_eval = df_eval.reset_index([0, 1, 2, 3])
```

```python
df_eval
```

```python
selection_metric = 'auc'
best_performance = df_eval.loc[(df_eval.metric == selection_metric) & (df_eval.phase == 'val')].groupby(['phase', 'outcome', 'sensitive_variable'])[['performance']].agg(max).merge(df_eval)
best_performance = df_eval.loc[(df_eval.metric == 'loss') & (df_eval.phase == 'val')]. \
    groupby(['phase', 'outcome'])[['performance']].agg(max).merge(df_eval)
best_performance
df_eval.merge(best_performance[['experiment_id', 'outcome', 'sensitive_variable']])
```

```python
selection_metric = 'loss'
best_performance = df_eval.loc[(df_eval.metric == selection_metric) & (df_eval.phase == 'val')].groupby(['phase', 'outcome', 'sensitive_variable'])[['performance']].agg(max).merge(df_eval)
best_performance = df_eval.loc[(df_eval.metric == 'loss') & (df_eval.phase == 'val')]. \
    groupby(['phase', 'outcome', 'sensitive_variable'])[['performance']].agg(min).merge(df_eval)
best_performance
df_eval.merge(best_performance[['experiment_id', 'outcome', 'sensitive_variable']])
```

```python
best_performance[['outcome', 'experiment_id', 'sensitive_variable']]
```

```python
best_model_ids = best_performance[['outcome', 'experiment_id', 'sensitive_variable']]. \
    set_index(['outcome', 'sensitive_variable']).transpose().to_dict('records')
```

```python
best_model_ids = best_model_ids[0]
print(best_model_ids)
```

```python
# ## Create defaults
config_path = os.path.join(data_path, 'config', 'grid', 'cfvae')
config_defaults_path = os.path.join(data_path, 'config', 'defaults', 'cfvae')
for the_key, the_model_id in best_model_ids.items():
    the_outcome, the_sensitive_variable = the_key
#     Read the best config
    with open(os.path.join(config_path, the_sensitive_variable, '{}.yaml'.format(the_model_id)), 'r') as fp:
        best_config = yaml.load(fp)
    temp_path = os.path.join(config_defaults_path, the_outcome, the_sensitive_variable)
    os.makedirs(temp_path, exist_ok = True)  
    with open(os.path.join(temp_path, 'model_config.yaml'), 'w') as fp:
        yaml.dump(best_config, fp)
```
